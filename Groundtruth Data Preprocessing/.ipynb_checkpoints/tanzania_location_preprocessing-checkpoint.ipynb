```python
import pandas as pd
from pandas import read_csv
```

```python
#Load the data file containing the household address,this file is generated from HH_SEC_A.csv 
location2013 =  read_csv('Tanzania/2012_2013/tanzania_2012_2013_preprocessed_data/tanzania_2012_2013_household_location.csv', header=0,delimiter=',')
location2013.head()
```

```python
#read the file containing the location of the household 
location2010 =  read_csv('Tanzania/2010_2011/tanzania_2010_2011_preprocessed_data/tanzania_2010_2011_household_location.csv', header=0,delimiter=',')
location2010.head()
```

```python
# Merge dataframe A and B based on 'region' and 'district_id'
merged_df = pd.merge(location2010, location2013[['region','district_id','district']], on=['region', 'district_id'], how='left')
merged_df.head()
```

```python
merged_df.shape
```

```python
merged_df = merged_df.drop_duplicates()
# Optionally, you can reset the index if you want
merged_df = merged_df.reset_index(drop=True)
merged_df.shape
```

```python
#save the file may be useful later
merged_df.to_csv('Tanzania/2010_2011/tanzania_2010_2011_preprocessed_data/tanzania_2010_2011_household_location_new.csv', index=False)
```

#### Concatenate datasets


```python
import pandas as pd

# Load each dataset into a separate dataframe
df1 = pd.read_csv('Tanzania/location_tz/1.csv')
df2 = pd.read_csv('Tanzania/location_tz/2.csv')
df3 = pd.read_csv('Tanzania/location_tz/3.csv')
df4 = pd.read_csv('Tanzania/location_tz/4.csv')
df5 = pd.read_csv('Tanzania/location_tz/5.csv')
df6 = pd.read_csv('Tanzania/location_tz/6.csv')
df7 = pd.read_csv('Tanzania/location_tz/7.csv')
# Repeat this step for each dataset you want to merge

# Merge datasets using the concat function
merged_df = pd.concat([df1, df2, df3, df4, df5, df6, df7], ignore_index=True)

# Alternatively, if you want to merge based on a common column, you can use the merge function
# merged_df = pd.merge(df1, df2, on='common_column', how='inner')
merged_df.head()

```

```python
merged_df.shape
```

```python
# Save the merged dataframe to a new CSV file
merged_df.to_csv('tanzania_merged_location.csv', index=False)
```
